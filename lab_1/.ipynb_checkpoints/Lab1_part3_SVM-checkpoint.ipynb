## 1.1 Pure models.
First let's try to solve this case with good old Logistic Regression and simple (linear kernel) SVM classifier.

Train LR and SVM classifiers (choose params by hand, no CV or intensive grid search neeeded) and plot their decision regions. Calculate one preffered classification metric.

Describe results in one-two sentences.

_Tip:_ to plot classifiers decisions you colud use either sklearn examples ([this](https://scikit-learn.org/stable/auto_examples/neural_networks/plot_mlp_alpha.html#sphx-glr-auto-examples-neural-networks-plot-mlp-alpha-py) or any other) and mess with matplotlib yourself or great [mlxtend](https://github.com/rasbt/mlxtend) package (see their examples for details)

_Pro Tip:_ wirte function `plot_decisions` taking a dataset and an estimator and plotting the results cause you want to use it several times below

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# from mlxtend.plotting import plot_decision_regions

lr = LogisticRegression(C=1, fit_intercept=True, max_iter=300, random_state=42) # add some params
svm = SVC(kernel='linear', C=1, max_iter=300, random_state=42) # here too

### YOUR CODE HERE
from matplotlib.colors import ListedColormap

X, y = moons_points, moons_labels
lr = lr.fit(X, y)
svm = svm.fit(X, y)
```

```python
def plot_decisions(dataset, estimator, title, poly_features=None):
    X, y = dataset
    cmap = ListedColormap(['r', 'g', 'b', 'yellow'])
    
    # fig, ax = plt.subplots(1, 1, figsize=(5, 5))
    if poly_features is None:
        x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
        y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
        # h = min((y_max - y_min) / 1000, (x_max - x_min) / 1000)
        h = 0.01
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
        plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap)
        Z = estimator.predict(np.c_[xx.ravel(), yy.ravel()])
    else:
        x_min, x_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
        y_min, y_max = X[:, 2].min() - 0.5, X[:, 2].max() + 0.5
        # h = min((y_max - y_min) / 1000, (x_max - x_min) / 1000)
        h = 0.01
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
        plt.scatter(X[:, 1], X[:, 2], c=y, cmap=cmap)
        Z = estimator.predict(poly_features.transform(np.c_[xx.ravel(), yy.ravel()]))

    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.25)
    # plt.contour(xx, yy, Z, colors='k', linewidths=0.7)

    plt.title(title)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())
    
```

```python
dataset = (X, y)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
plt.subplot(121)
plot_decisions(dataset=dataset, estimator=lr, title='Decision region for log reg')
plt.subplot(122)
plot_decisions(dataset=dataset, estimator=svm, title='Decision region for SVM')
```

```python
from sklearn.metrics import roc_auc_score

print('ROC AUC for log reg :', roc_auc_score(y, lr.predict(X)))
print('ROC AUC for SVM with linear kernel :', roc_auc_score(y, svm.predict(X)))
```

## 1.2 Kernel tirck

Now use different kernels (`poly`, `rbf`, `sigmoid`) on SVC to get better results. Play `degree` parameter and others.

For each kernel estimate optimal params, plot decision regions, calculate metric you've chosen eariler.

Write couple of sentences on:

* What have happenned with classification quality?
    - 'rbf' ядро показало себя лучше всего и по метрике ROC-AUC, и по визуальному восприятию decision region для SVM,в этой задаче хочется использовать именно его; у 'poly' качество разнилось для разных степеней полиномиального ядра(лучшая степень -- 3, но 'rbf' не догоняет даже чисто визуально, по ROC-AUC тоже), 'sigmoid' не выглядит хорошим вариантом, много ошибок классификации
* How did decision border changed for each kernel?
    - у полиномиального ядра сильно меняется decision region при разных степенях, при четных степенях у одного из классов появляется две непересекающиеся области
* What `degree` have you chosen and why?
    - лучшее качество было получено для степени 3 для ядра 'poly', также у нее наиболее адекватный decision region

```python
%%time
### YOUR CODE HERE

fig, ax = plt.subplots(9, 3, figsize=(12, 27))
for plots_column, kernel_name in enumerate(['poly', 'rbf', 'sigmoid']):
    for degree in range(1, 10):
        svm = SVC(
            C=1.0, kernel=kernel_name, degree=degree, random_state=42
        ).fit(X, y)
        
        plt.subplot(9, 3, (degree - 1) * 3 + plots_column + 1)
        plot_decisions(
            dataset, svm, 
            f'SVM, {kernel_name}, deg={degree}, ROC-AUC={np.round(roc_auc_score(y, svm.predict(X)), decimals=2)}'
        )
        
```

## 1.3 Simpler solution (of a kind)
What if we could use Logisitc Regression to successfully solve this task?

Feature generation is a thing to help here. Different techniques of feature generation are used in real life, couple of them will be covered in additional lectures.

In particular case simple `PolynomialFeatures` ([link](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html)) are able to save the day.

Generate the set of new features, train LR on it, plot decision regions, calculate metric.

* Comare SVM's results with this solution (quality, borders type)
    - качество с полиномиальными фичами до старшей нечетной получилось практически таким же, как у SVM с 'rbf' ядром (по визуальному восприятию, а также по метрике ROC-AUC)
    - границы получились почти одинаковые для LR с нечетным poly у PolynomialFeatures, очень похожи на SVM с 'rbf' ядром
* What degree of PolynomialFeatures have you used? Compare with same SVM kernel parameter.
    - посмотрел на степени от 2 до 9, между нечетными почти никакой разницы, но наибольший ROC-AUC при poly=7, его и стоит выбрать, по моему мнению

```python
%%time
from sklearn.preprocessing import PolynomialFeatures

### YOUR CODE HERE
fig, ax = plt.subplots(4, 2, figsize = (10, 10))

for poly in range(2, 10):
    plt.subplot(4, 2, poly - 1)
    
    poly_feat = PolynomialFeatures(poly)
    poly_X = poly_feat.fit_transform(X)
    poly_dataset = poly_X, y
    
    lr = LogisticRegression(C=1, fit_intercept=True, random_state=42, max_iter=500)
    lr = lr.fit(poly_X, y)
    
    plot_decisions(
        poly_dataset, lr, 
        f'LR, poly={poly}, ROC-AUC={roc_auc_score(y, lr.predict(poly_X))}', 
        poly_features=poly_feat
    )
```

```python
from sklearn.metrics import f1_score, accuracy_score
```

```python
### YOUR CODE HERE

# simple LR
lr = LogisticRegression(C=1, fit_intercept=True, random_state=42, max_iter=500)
lr = lr.fit(points, labels)
print('F1 weighted for simple LR :', f1_score(labels, lr.predict(points), average='weighted'))
print('Accuracy for simple LR :', accuracy_score(labels, lr.predict(points)))
plot_decisions((points, labels), lr, 'simple LR for 4 class clf')
```

Простая логистическая регрессия показала себя здесь плохо, что неуидивительно: данные довольно сложные, модель простая, данные никак не предобработаны (нет генерации новых фичей и т.д.)

```python
from scikitplot.metrics import plot_roc

plot_roc(labels, lr.predict_proba(points))
```

```python
from warnings import filterwarnings
filterwarnings(action='ignore')
```

```python
## Log Reg with polynomial features

fig, ax = plt.subplots(4, 2, figsize = (10, 10))

for poly in range(2, 10):
    plt.subplot(4, 2, poly - 1)
    
    poly_feat = PolynomialFeatures(poly)
    poly_points = poly_feat.fit_transform(points)
    poly_dataset = poly_points, labels
    
    lr = LogisticRegression(C=1, fit_intercept=True, random_state=42, max_iter=500)
    lr = lr.fit(poly_points, labels)
    
    plot_decisions(
        poly_dataset, lr, 
        f'LR, poly={poly}, ACC={accuracy_score(labels, lr.predict(poly_points))}', 
        poly_features=poly_feat
    )
```

Логистическая регрессия с полиномиальными фичами обладает куда большим качество и визуально, и по метрике accuracy. При этом лучшее качество, на мой взгляд, получилось при poly=6 (оцениввал визуально и по метрике accuracy качество все же тоже выше).

```python
# SVM with 'rbf' kernel, degree from 2 to 9

svm = SVC(
    C=1.0, kernel='rbf', random_state=42, probability=True
).fit(points, labels)

plot_decisions(
    (points, labels), svm, 
    f'SVM, rbf, ACC={accuracy_score(labels, svm.predict(points))}'
)
```

```python
plot_roc(labels, svm.predict_proba(points))
```

SVM с 'rbf' ядром показал хорошее качество даже на непредобработанных (без генерации фичей и т.д.) данных : визуально и по accuracy. 

Однако можно заметить, что синиже и желтые точки он классифицировал неидеально, также границы областей на стыке желтой, синей и зеленой областей получились слишком угловатые, это сложно интерпретировать.

```python
# try XGBClassifier

import xgboost as xgb
from xgboost import XGBClassifier

fid, ax = plt.subplots(4, 2, figsize=(10, 20))

for depth in range(2, 10):
    xgbc = xgb.XGBClassifier(
        n_estimators=200, 
        max_depth=depth, 
        learning_rate=1e-3, 
        n_jobs=-1,
        random_state=42
    ).fit(points, labels)
    
    plt.subplot(4, 2, depth - 1)
    plot_decisions((points, labels), xgbc, f'XGB classifier, depth={depth}, ACC = {accuracy_score(labels, xgbc.predict(points))}')
```

```python
xgbc = xgb.XGBClassifier(
    n_estimators=200, 
    max_depth=9, 
    learning_rate=1e-3, 
    n_jobs=-1,
    random_state=42
).fit(points, labels)

plot_roc(labels, xgbc.predict_proba(points))
```

XGB бустинг из 200 деревьев глубины 9 имеет самую большую accuracy из всех моделей, которые были использованы, визуально разбиение на области классов выглядит логичным, но есть и огрехи: зеленая и синия области имеют немного странную границу, логично было бы ее провести немного выше. Но, возможно, просто недостаточно глубокие деревья использовались в ансамбле.

Проверим глубину 10-17:

```python
fid, ax = plt.subplots(4, 2, figsize=(10, 20))

for depth in range(10, 18):
    del xgbc
    xgbc = xgb.XGBClassifier(
        n_estimators=200, 
        max_depth=depth, 
        learning_rate=1e-3, 
        n_jobs=-1,
        random_state=42
    ).fit(points, labels)
    
    plt.subplot(4, 2, depth - 9)
    plot_decisions((points, labels), xgbc, f'XGB classifier, depth={depth}, ACC = {accuracy_score(labels, xgbc.predict(points))}')
```

Видно, что увеличивать глубину после 10 уже не имеет смысла: данных не так много, поэтому при увеличении глубины деревьев получим лишь сильноее переобучение, но не прирост качества.
