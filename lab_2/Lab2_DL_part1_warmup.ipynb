````python
# Iptimizer params
optimizer_config = {'learning_rate' : 1e-1, 'momentum': 0.9}
optimizer_state = {}

# Looping params
n_epoch = 20
batch_size = 128
````

````python
# batch generator
def get_batches(dataset, batch_size):
    X, Y = dataset
    n_samples = X.shape[0]
        
    # Shuffle at the start of epoch
    indices = np.arange(n_samples)
    np.random.shuffle(indices)
    
    for start in range(0, n_samples, batch_size):
        end = min(start + batch_size, n_samples)
        
        batch_idx = indices[start:end]
    
        yield X[batch_idx], Y[batch_idx]
````

````python
# Your code goes here. ################################################

# compare activation functions -- losses and training time
from time import time
from tqdm import tqdm

epochs = 40
batch_size = 96

criterion = ClassNLLCriterion()

opt = sgd_momentum
opt_conf, opt_state = {'learning_rate' : 1e-2, 'momentum': 0.9}, {}


def get_model(activation_name, use_batch_norm=False):
    all_activation_names = {'ReLU', 'ELU', 'LeakyReLU', 'SoftPlus'}
    if activation_name not in all_activation_names:
        raise f"Activation name {activation_name} isn't allowed ... Good names are: {all_activation_names}"
    else:
        if activation_name == 'ReLU':
            activation = ReLU()
        elif activation_name == 'ELU':
            activation = ELU()
        elif activation_name == 'LeakyReLU':
            activation = LeakyReLU()
        elif activation_name == 'SoftPlus':
            activation = SoftPlus()
    
    model = Sequential()
    
    model.add(Linear(n_in=n_in, n_out=40))
    if use_batch_norm:
        model.add(BatchNormalization(alpha=0.9))
    model.add(activation)
    
    model.add(Linear(n_in=40, n_out=40))
    if use_batch_norm:
        model.add(BatchNormalization(alpha=0.9))
    model.add(activation)
    
    model.add(Linear(n_in=40, n_out=40))
    if use_batch_norm:
        model.add(BatchNormalization(alpha=0.9))
    model.add(activation)
    
    model.add(Linear(n_in=40, n_out=n_out))
    model.add(LogSoftMax())
    return model

    
    
def train_eval_models(models, opt=opt, opt_conf=opt_conf, opt_state=opt_state):
    model_losses_train = {'ReLU': None, 'ELU': None, 'LeakyReLU': None, 'SoftPlus': None}
    model_losses_val = {'ReLU': None, 'ELU': None, 'LeakyReLU': None, 'SoftPlus': None}
    train_times = []
    
    for (m_num, m), act in zip(enumerate(models), ['ReLU', 'ELU', 'LeakyReLU', 'SoftPlus']):
        train_losses = [] # for each epoch
        val_losses = [] # for each epoch
        train_time = 0.0
    
        pbar = tqdm(range(epochs), desc='Total')
        for i in pbar:
            pbar.set_description(
                f'Model {m_num+1}/{len(models)}, Loss: ' + \
                f'{0.0 if len(train_losses) == 0 else train_losses[-1]}'
            )
            
            m.train()
            train_loss = 0.0
            start = time()
            for x_batch, y_batch in get_batches((X_train, y_train), batch_size):
                m.zeroGradParameters()
                probs = m.forward(x_batch)                
                loss = criterion.forward(probs, y_batch)
                dp = criterion.backward(probs, y_batch)
                m.backward(x_batch, dp)
                opt(m.getParameters(), m.getGradParameters(), 
                    opt_conf, opt_state
                )
                train_loss += loss * x_batch.shape[0] / X_train.shape[0]
            end = time()
            train_losses.append(train_loss)
            train_time += end - start
        
            m.evaluate()
            val_loss = 0.0
            for x_batch, y_batch in get_batches((X_val, y_val), batch_size):
                probs = m.forward(x_batch)
                loss = criterion.forward(probs, y_batch)
                val_loss += loss * x_batch.shape[0] / X_val.shape[0]
            val_losses.append(val_loss)
        
        model_losses_train[act] = train_losses
        model_losses_val[act] = val_losses
        train_times.append(train_time)
    
    return model_losses_train, model_losses_val, train_times
````

````python
def compare_activations(model_losses_train, model_losses_val, train_times, title):
    fig, axes = plt.subplots(2, 2, figsize=(10, 10), sharey=True)
    
    plt.suptitle(title)

    plt.ylabel('loss')
    plt.yscale('log')
    plt.ylim(5e-3, 1.0)
    
    plt.subplot(221)
    plt.title('Train loss')
    plt.xlabel('epoch')
    plt.plot(range(epochs), model_losses_train['ReLU'], label='ReLU')
    plt.plot(range(epochs), model_losses_train['ELU'], label='ELU')
    plt.plot(range(epochs), model_losses_train['LeakyReLU'], label='LeakyReLU')
    plt.plot(range(epochs), model_losses_train['SoftPlus'], label='SoftPlus')
    plt.grid(True)
    plt.legend()

    plt.subplot(222)
    plt.title('Val loss')
    plt.xlabel('epoch')
    plt.plot(range(epochs), model_losses_val['ReLU'], label='ReLU')
    plt.plot(range(epochs), model_losses_val['ELU'], label='ELU')
    plt.plot(range(epochs), model_losses_val['LeakyReLU'], label='LeakyReLU')
    plt.plot(range(epochs), model_losses_val['SoftPlus'], label='SoftPlus')
    plt.grid(True)
    plt.legend()

    plt.subplot(212)
    plt.title('Training time')
    plt.xlabel('activation')
    plt.ylabel('time, s')
    plt.bar(['ReLU', 'ELU', 'LeakyReLU', 'SoftPlus'], train_times)
    
    plt.tight_layout()
    plt.show()
````
